```python
def construct_A(N, num_nodes):
    # Construct Basic Matrix of ranges 
    A_temp = []
    for i in range(N):
        for j in range(i,N):
            temp = [0 for _ in range(num_nodes)]
            for k in range(i,j+1):
                temp[k] = 1
            A_temp.append(temp)
    A_temp = matrix(A_temp)

    # Reorder matrix and v_list
    temp_size = num_nodes
    A = []
    
    # The rows with elementary volumes
    if N == 5:
        nodes = [0,5,9,12,14]
    elif N == 10:
        nodes = [0,10,19,27,34,40,45,49,52,54]
    elif N == 12:
        nodes = [0,12,23,33,42,50,57,63,68,72,75,77]
    elif N == 15:
        nodes = [0,15,29,42,54,65,75,84,92,99,105,110,114,117,119]
        
    all_nodes = [i for i in range(num_nodes)]
    set_nodes = set(nodes)
    set_all_nodes = set(all_nodes)
    diff_nodes = list(set_all_nodes - set_nodes)

    for i in nodes:
        A.append(list(list(A_temp[i,:temp_size])[0]))
    for i in range(temp_size-N):
        A.append(list(list(A_temp[diff_nodes[i],:temp_size])[0]))

    big_num = 10**6
    idx_big_num = N
    for i in range(N,temp_size):
        A[i][idx_big_num] = big_num
        idx_big_num += 1
    A = matrix(A)

    return A


def get_closest_vols(val, vols):
    min_dist = max(vols)
    res = 0
    for i in vols:
        ratio = abs(float((i-val)/i))
        if ratio < min_dist:
            min_dist = ratio
            res = i
    return int ((1.001) * res)


def get_vlist(elem_vols, vols):
    res = []
    for i in range(N):
        for j in range(i,N):
            temp = 0
            for k in range(i,j+1):
                temp += elem_vols[k]
            res.append(get_closest_vols(temp, vols))   
    return res


def construct_v(v_list, N, num_nodes):
    v = []
    for val in v_list:
        v.append(int(val))
    v = vector(v)

    temp_size = num_nodes

    if N == 5:
        nodes = [0,5,9,12,14]
    elif N == 10:
        nodes = [0,10,19,27,34,40,45,49,52,54]
    elif N == 12:
        nodes = [0,12,23,33,42,50,57,63,68,72,75,77]
    elif N == 15:
        nodes = [0,15,29,42,54,65,75,84,92,99,105,110,114,117,119]

    all_nodes = [i for i in range(num_nodes)]
    set_nodes = set(nodes)
    set_all_nodes = set(all_nodes)
    diff_nodes = list(set_all_nodes - set_nodes)
    v_ordered = []
    for i in nodes:
        v_ordered.append(v[i])

    for i in range(temp_size-N):
        v_ordered.append(v[diff_nodes[i]])

    return v_ordered


# Given an array of size n, computes the volumes of all the ranges
# output sizeL n(n-1)/2 + n
def range_computer(arr, with_noise = False):
    n = len(arr)
    # print('input length:', n)

    output = []
    sum = 0
    for i in range(n):
        sum = arr[i]
        output.append(sum)
        if with_noise:
            output.append(sum-1)
            output.append(sum+1)
        for j in range(1,n-i):
            sum += arr[i+j]
            output.append(sum)
            if with_noise:
                output.append(sum-1)
                output.append(sum+1)
    output.sort()

    # print('ouput length:', len(output))
    return output

def sol_diff(real, guess):
    a = np.array(real)
    b = np.array(guess)
    c = (a - b) / b
    return vector(c).norm()
```
